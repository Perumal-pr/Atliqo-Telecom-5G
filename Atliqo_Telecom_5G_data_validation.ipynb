```python
# Loading the Datasets into sql using Mysql+pymysql

from sqlalchemy import create_engine
from urllib.parse import quote_plus

host="localhost"
port="3306"
database="telecom"
user="root"
password= quote_plus("Password@123")

engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")

metrics="fact_atliqo_metrics"
f_metrics.to_sql(metrics,engine, if_exists="replace",index=False)

print(f"Successfully loaded the table {metrics} in database {database}.")
```
